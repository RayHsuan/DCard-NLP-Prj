```python
#%%time
import pandas as pd
import matplotlib.pyplot as plt
import string
import math

tmp = open('C:/Data/PA-2/stopword_chinese.txt','r',encoding = 'utf-8')
stopword = list(tmp.read().split('\n'))
tmp.close()
tmp = open('C:/Data/PA-2/punctuation.txt','r',encoding = 'utf-8') #.read().split('\n')
punctuation = list(tmp.read().split('\n'))
tmp.close()
#print(punctuation)
for i, val in enumerate(punctuation):
    if u'\ufeff' in val:
        t = val.split('.')
        punctuation[i] = '.'

#print(punctuation)    
RemovedWords = stopword + punctuation
print(RemovedWords)
#print(RemovedWords)
#\s         | 空格、定位符號或換行符號(即比對空白字元)

#\S         | 除了空格、定位符號和換行符號以外的任何字元
#movieRegex = re.compile(r'[^funy]') #比對是否有f、u、n、y以外的字元
#movieRegex.findall('See you tomorrow.')
```

```python
%%time
tmp = open('C:/Data/PA-2/TMBD_news_files_2000.txt', 'r',encoding = 'utf-8')
docs = list(tmp.read().split('\n'))
tmp.close()
lstToken = []
lstDoc = []
dictDoc = {}

for i, line in enumerate(docs):
    #Get Doc ID
    ID = int(line.split('\t')[0]) #依照文檔讀取每一列Doc ID
    #Get Split Terms
    content = line.split('\t')[1].strip() #依照文檔讀取每一列的文章斷詞
    content2 = ""
    #print(ID)
    #print(content)

    for tokens in content.split():
        if tokens.strip().lower() not in RemovedWords:
            #if tokens.strip().lower() not in lstToken:
            lstToken.append(tokens.strip().lower())
                
            content2 += tokens.strip().lower() + " "
    #print(content2)
    #Put ID and Content Pair in Dictionary
    #dictDoc[ID] = content2 #將 ID 和 所有斷詞記錄在 Dict中
    dictDoc[ID] = content2
    lstDoc.append(content2)

    #if ID == 1961:
    #    print(content.split())
    #if ID == 2000:
        #print(content.split())

#print(len(dictDoc))
#print(dictDoc[1])
#print(dictDoc[2000])
print(len(lstToken))
print(len(dictDoc))
print(len(lstDoc))
```

```python
%%time
Term_freq = {}
lbldf = ["Term","TF","DF","IDF","Weight"]
dfStep1 = pd.DataFrame(columns = lbldf)
#老師有交代，如果TF太低的字都可以剔除，減少維度
for Term in lstToken:
    if Term in Term_freq:
        Term_freq[Term] = Term_freq[Term] + 1
    else:
        Term_freq[Term] = 1
        
#print(len(Term_freq))
for key in Term_freq:
    lstStep1 = []
    df = 0
    #print(key)
    #print(Term_freq[key])
    for ID in dictDoc:
        if key in dictDoc[ID]:
            df+=1
    if df == 0:
        print(key)
    
    lstStep1.append(key)
    lstStep1.append(Term_freq[key])
    lstStep1.append(df)
    lstStep1.append(math.log(len(dictDoc)/df))
    lstStep1.append(Term_freq[key]*(math.log((len(dictDoc)/df))))
    ##dfStep1.loc[dfStep1.shape[0]] = lstStep1
    dfStep1.loc[len(dfStep1.index)] = lstStep1

print(len(dfStep1.columns))
print(len(dfStep1))  
#Term_freq        
#dfStep1
```

```python
%%time
import numpy as np
tfidf_mx = []
def tf_idf(length,document):
    tfidf_v = np.zeros([length,1])
    for term in document.split():
        tfidf_v[vocabulary.index(term),0] = Weight[vocabulary.index(term)]
    return tfidf_v
    

for doc in lstDoc:
    tfidf = tf_idf(len(vocabulary),doc)
    #print(tfidf_mx)
    tfidf_mx.append(tfidf)
    if lstDoc.index(doc) % 50 == 0:
        print(lstDoc.index(doc))
    #if doc == lstDoc[2]:
    #    break
    
```

```python
from numpy import linalg as LA
dictSim = {}
def getSimilarity(ref_v,compare_v):
    return (np.dot(ref_v.T,compare_v)[0][0]) / (LA.norm(ref_v) * LA.norm(compare_v))

j = 0
for i, vec in enumerate(tfidf_mx):
    j+=1
    dictSim[j] = getSimilarity(tfidf_mx[55],tfidf_mx[i])
```
